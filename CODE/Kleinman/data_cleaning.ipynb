## Group 5 Data Cleaning for Hennepin County Analysis

```python
#Code Block 1
## import necessary libraries

import pandas as pd
import geopandas as gpd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
import os
import sys

plt.style.use("seaborn-v0_8")
sns.set_style("whitegrid")

%matplotlib inline
```

```python
#Code Block 2
## load in the data

data_pull = pd.read_csv('../../Data/Hennepin/mortgage_covenant_data-new.csv')
data_given = pd.read_csv('../../Data/mortgage_covenant_data.csv')

display(data_pull.info())
data_given.info()
```

```python
#Code Block 3
## view the head of the data
display(data_pull.head())
data_given.head()
```

```python
#Code Block 4
## view the missing values

print("Pulled Data Missing Values:")
display(data_pull.isnull().sum())

print("Given Data Missing Values:")
data_given.isnull().sum()
```

```python
#Code Block 5
## fix data types

data_pull['property_value'] = pd.to_numeric(data_pull['property_value'], errors='coerce')
data_pull['interest_rate'] = pd.to_numeric(data_pull['interest_rate'], errors='coerce')

display(data_pull.info(), data_given.info())
```

```python
#Code Block 6
## census tract formatting
print(type(data_pull['census_tract'][0]))

### convert to string to merge tract area for covenant density
data_pull['census_tract'].head()
```

```python
#Code Block 7
## handle missing values
### drop rows with missing census data
data_pull = data_pull.dropna(subset=['census_tract'])

### fill nulls for property value and income
data_pull['property_value'] = data_pull['property_value'].fillna(data_pull['property_value'].median())
data_pull['income'] = data_pull['income'].fillna(data_pull['income'].median())

display(data_pull.info())
data_pull.isnull().sum()
```

```python
#Code Block 8 
## recreate covenant density feature
display(data_given['covenant_density'].value_counts())

## recreate covenant density feature for pulled data based on covenant count
bins = [0, 50, 100, 200, float('inf')]
labels = ['Low (1-50)', 'Medium (51-100)', 'High (101-200)', 'Very High (>200)']
data_pull['covenant_density'] = pd.cut(data_pull['covenant_count'], bins=bins, labels=labels, right=True)

# Verify the value counts for the new covenant_density column
display(data_pull['covenant_density'].value_counts())

### diagnostics
display(data_pull.info())
data_pull.isnull().sum()
```

```python
#Code Block 9
## fill the missing values for covenant density with 0 (tracts with no covenants)

## Add 0 as a category to the covenant_density column
data_pull['covenant_density'] = data_pull['covenant_density'].cat.add_categories([0])

## Fill the missing values for covenant density with 0 (tracts with no covenants)
data_pull['covenant_density'] = data_pull['covenant_density'].fillna(0)

## Check for remaining missing values
data_pull.isnull().sum()
```

```python
#Code Block 10
## do the same for the given data

display(data_given.info())
data_pull.info()
```

```python
#Code Block 11
## fixing the given data

### convert and reformat census tract string
data_given['census_tract'] = (
    data_given['census_tract']
    .round()
    .astype('Int64')
    .astype(str)
    .str.zfill(11)
)

### drop rows with missing census data
data_given = data_given.dropna(subset=['census_tract'])

### fill nulls for property value and income
data_given['property_value'] = data_given['property_value'].fillna(data_given['property_value'].median())
data_given['income'] = data_given['income'].fillna(data_given['income'].median())

### fill nulls for covenant density with 0
data_given['covenant_density'] = data_given['covenant_density'].fillna(0)

display(data_given.info())
data_given.isnull().sum()
```

```python
#Code Block 12
## save the cleaned datasets
if not os.path.exists('../../Data/Hennepin/cleaned'):
    os.makedirs('../../Data/Hennepin/cleaned')

data_pull.to_csv('../../Data/Hennepin/cleaned/cleaned_mortgage_covenant_data_pull.csv', index = False)
data_given.to_csv('../../Data/Hennepin/cleaned/cleaned_mortgage_covenant_data.csv', index = False)
```
